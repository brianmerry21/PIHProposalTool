```python
import openpyxl
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, PageBreak, Paragraph, Spacer
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER


# Load Excel file
wb = openpyxl.load_workbook("C:/Users/Deftsoft-Python/Documents/PdfDataExtractor (1)/PdfDataExtractor/static/uploads/eec91127-d297-4878-bba9-10c08328b588.xlsx")
ws = wb.active  # or specify sheet name ws = wb["Sheet1"]

# Extract data from B35:B75, C35:C75, and K35:K75
data = []
for row in range(37, 76):
    b_val = ws[f"B{row}"].value
    c_val = ws[f"C{row}"].value
    k_val = ws[f"K{row}"].value

    # Skip blank rows if there's a space
    if b_val is None and c_val is None and k_val is None:
        continue

    data.append([b_val, c_val, k_val])

# Create PDF
pdf_file = "output.pdf"
doc = SimpleDocTemplate(pdf_file, pagesize=A4)

# Add dummy pages until page 9
story = []
for _ in range(8):  
    story.append(PageBreak())
    
# <-- Ensure styles is defined -->
styles = getSampleStyleSheet()

# Custom centered + large heading style
centered_heading = ParagraphStyle(
    "CenteredHeading",
    parent=styles["Heading1"],   # Heading1 is already bigger than Heading2
    alignment=TA_CENTER,         # Center text
    fontSize=18,                 # Make it larger
    leading=22,                  # Line spacing for better look
    spaceAfter=20,               # Add space after heading
    textColor=colors.HexColor("#000000"),  # Optional: heading color
)


# push content closer to bottom
story.append(Spacer(1,200)) 

# Adding the heading    
heading = Paragraph("Optional Items", centered_heading)
story.append(heading)
story.append(Spacer(1, 15))

# Create table
table = Table([["Items", "Price Est.", "Qty"]] + data, repeatRows=1)
table.setStyle(TableStyle([
    ("BACKGROUND", (0,0), (-1,0), colors.lightgrey),
    ("GRID", (0,0), (-1,-1), 1, colors.black),
    ("ALIGN", (0,0), (-1,-1), "CENTER"),
    ("VALIGN", (0,0), (-1,-1), "MIDDLE"),
]))

story.append(table)

# Build PDF
doc.build(story)

print("✅ Data extracted and placed on page 9 of PDF.")
```
